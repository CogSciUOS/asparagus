```python
import numpy as np
import random
import scipy.stats as stats
import matplotlib.pyplot as plt

''' Dummy method. Returns coordinates either for a straight or bended piece randomly'''
def aspargus_slices(preprocessed_image, y_pos_of_slices):
    if(random.randint(0, 1)):
        return [[100,110],[100,110],[100,110]]
    else:
        return [[100,110],[110,115],[100,110]]

""" Returns a score for the curvature of the aparagus piece. A perfectly straight aspargus yields a score of 0"""
def curvature_score(preprocessed_image):
    y_pos_of_slices = [100,120,180]#or whatever...
    centers = np.mean(np.array(aspargus_slices(preprocessed_image, y_pos_of_slices)),axis=1)
    plt.scatter(y_pos_of_slices,centers)   
    slope, intercept, r_value, p_value, std_err = stats.linregress(y_pos_of_slices,centers)
    return std_err
```

```python
x = curvature_score(42)
print(x)
```

```python
def fig2rgb_array(fig):
    fig.canvas.draw()
    buf = fig.canvas.tostring_rgb()
    ncols, nrows = fig.canvas.get_width_height()
    return np.frombuffer(buf, dtype=np.uint8).reshape(nrows, ncols, 3)

fig2, ax = plt.subplots(figsize=(20,20))
ax.imshow(fig2rgb_array(fig))
```
